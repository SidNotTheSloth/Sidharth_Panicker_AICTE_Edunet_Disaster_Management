Project Title

Wildfire Risk Prediction using Climate & Vegetation Drivers

Problem Statement

Wildfires are becoming more frequent and severe due to climate change, prolonged droughts, and increased vegetation flammability. Predicting whether a wildfire will ignite (Yes/No) based on factors like temperature, precipitation, wind speed, vegetation, and historical trends can enhance early warning systems and improve wildfire management strategies.

Dataset

California Weather and Fire Prediction Dataset (1984â€“2025) with Engineered Features (Zenodo)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/CA_Weather_Fire_Dataset_1984-2025.csv')

print(df.head(), "\n")

print(df.info(), "\n")

print(df.describe(), "\n")

print(df.isnull().sum(), "\n")

#Visualization
if 'FIRE_START_DAY' in df.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x='FIRE_START_DAY', data=df, palette='Set2')
    plt.title("Wildfire Ignition Distribution (Target Variable)")
    plt.show()

# Correlation heatmap of numerical features
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap of Climate Features")
plt.show()
```
